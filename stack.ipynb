Stack datastructure : it is a linear datastructure which follows the principle of LIFO, it follows the principle of last in first out. 
# common operations on stack
1. Push()
2. Pop()
3. IsEmpty()
4. IsFull()
5. Peek()
6. delete()

```python
class array:
    def __init__(self,length):
        self.array = []
        self.maxlength = length
        self.size = 0

    def __str__(self):
        return "Array Implementation"
    
    def get_array(self):
        return self.array

    def add_element(self,data):
        if self.is_full():
            # print("Array is Full")
            return False

        else:
            if self.size == 0:
                self.array.append(data)
                self.size+=1
                return True

            else:
                if type(data) == type(self.array[0]):
                    self.array.append(data)
                    self.size+=1
                    return True

                else:
                    print("Data type not Matching")
                    return False


    def is_full(self):
        if self.size == self.maxlength:
            return True
        else:
            return False
        
    def remove_by_index(self,index):
        if self.size>0:
            if index<self.size:
                self.array.pop(index)
                self.size-=1
                return True
            else:
                print("Index out of Range")
                return False
        else:
            # print("Array is Empty")
            return False

        
    def remove_element(self,data):
        if self.size>0:
            self.array.remove(data)
            self.size-=1
            print("Element removed successfully!!!")
            return True

        else:
            # print("Array is Empty")
            return False

    def display(self):
        print(self.array)

    def find_element(self,element):
        if element in self.array:
            print("Element Found")
            return self.array.index(element)
        
        else:
            print("Element Not found")
            return False
        
    def update_element(self,index,element):
        if self.size-1<index:
            print("Index out of Range")
            return False
        else:
            self.array[index] = element
            print("Array updated!!!")
            return True
```

```python
"""1. Push()
2. Pop()
3. IsEmpty()
4. IsFull()
5. Peek()
6. delete()
"""

# Implementation of stack datastructure :
class stack:
    def __init__(self,length):
        self.stack = array(length)

    def push(self,element):
        result = self.stack.add_element(element)
        if result:
            print("Element added to stack")

        else:
            print("Element not added")

    def pop(self):
        result = self.stack.remove_by_index(self.stack.size-1)
        if result:
            print("Element removed from the stack")

        else:
            print("Element not removed!!!")

    def display_stack(self):
        self.stack.display()

    def Is_Empty(self):
        if self.stack.size == 0:
            return True
        else:
            return False
        
    def Is_Full(self):
        self.stack.is_full()

    def peek(self):
        stack = self.stack.get_array()
        if self.Is_Empty():
            print("Stack is Empty!!!")
        else:    
            return stack[-1]
    
    def delete(self):
        stack = self.stack.get_array()
        if self.Is_Empty():
            print("Stack is Empty!!!")

        else:
            while self.Is_Empty == False:
                self.pop(self.stack.size-1)
    
        
```
